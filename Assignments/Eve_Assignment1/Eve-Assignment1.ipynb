```python
import random
import time

  Start_time = time.time()
  
  def stableMatching(n, group1Preferences, group2Preferences):
    # Initially, all n group teams are unmatched
    unmatchedGroup1 = list(range(n))
    # None of the team in group1 has a match team yet, we denote this by the value None
    group1Teammatch = [None] * n                      
    # None of the team in group2 has a match team yet, we denote this by the value None
    group2Teammatch = [None] * n                      
    # Each group1 team made 0 proposals, which means that 
    # its next proposal will be to the group2 team number 0 in its list
    nextg1teamChoice = [0] * n                       
    
    # While there exists at least one unmatched group1 team:
    while unmatchedGroup1:
        # Pick an arbitrary unmatched group1 team
        g1team = unmatchedGroup1[0]                      
        # Store its ranking in this variable for convenience
        g1teamPreferences = group1Preferences[g1team]       
        # Find a group2 team to propose to
        g2team = g1teamPreferences[nextg1teamChoice[g1team]] 
        # Store group2 team's ranking in this variable for convenience
        g2teamPreferences = group2Preferences[g2team]
        # Find the present match team of the selected group2 team (it might be None)
        currentMatch = group2Teammatch[g2team]      
        
        # Now "g1team" proposes to "g2team". 
        # Decide whether "g2team" accepts, and update the following fields
        # 1. group1Teammatch 
        # 2. group2Teammatch 
        # 3. unmatchedGroup1
        # 4. nextg1teamChoice
        if currentMatch == None:
          #No Match team case
          #"g2team" accepts any proposal
          group2Teammatch[g2team] = g1team
          group1Teammatch[g1team] = g2team
          #g1team nextchoice is the next g2team
          #in the giteamPreferences list
          nextg1teamChoice[g1team] = nextg1teamChoice[g1team] + 1
          #Delete "g1team" from the Unmatched list
          unmatchedGroup1.pop(0)


        else:
          #Matched team exists
          #Check the preferences of the 
          #current g1team and that of the proposed g1team's
          currentIndex = g2teamPreferences.index(currentMatch)
          g1teamIndex = g2teamPreferences.index(g1team)
          #Accept the proposal if 
          #"g1team" has higher preference in the g2teamPreference list
          if currentIndex > g1teamIndex:
             #New stable match is found for "g2team"
             group2Teammatch[g2team] = g1team
             group1Teammatch[g1team] = g2team
             nextg1teamChoice[g1team] = nextg1teamChoice[g1team] + 1
             #Pop the newly match team
             unmatchedGroup1.pop(0)
             #Now the previous matched team is unmatched add
             #g1team to the unmatched list
             unmatchedGroup1.insert(0,currentMatch)

          else:
             nextg1teamChoice[g1team] = nextg1teamChoice[g1team] + 1
    
    executionTime = time.time() - Start_time
    print("Execution time", executionTime)                                
  
    return group1Teammatch
    
# Complexity Upper Bound : O(n^2)
```

```python
# Calculate number of stable matchings/Total number(n)
# Check if stable after create new matching, if so, count+1  

# How to check whether is stable??
# Pseudocode about Stable-checker are as below
#   for every woman (w):
#      manList = a list of man that A prefers him over B
#      for every man (m) in manList:
#            if m prefers w over his current partner:
#               return false
# return stable                                                                                                                                     
def isStable(n, matching, g1p, g2p):
  # w is from group 1
  for w in range(n):  # w from 0 to n-1
    # w's current partner
    currentM = matching[w]
    # w's preference list
    wPreferences = g1p[w]
    # ranking(index) of currentM in w's preferences
    indexOfcurrentM = wPreferences.index(currentM)

    # j is every index < indexOfcurrentM, meaning ranks before currentM
    for j in range(0, indexOfcurrentM):
      # The man that's j-th in w's preferences
      m = wPreferences[j]
      # m's current partner (from group 1)
      currentP = matching.index(m)
      if g2p[m].index(w) < g2p[m].index(currentP):
        return False
  return True

def calculateStableMatchPercentage(n, group1Preferences, group2Preferences):
    countStableMatches = 0
    for i in range(1000):
        # Shuffle all group1Preferences
        for g1p in group1Preferences:
            random.shuffle(g1p)
        # Shuffle all group2Preferences
        for g2p in group2Preferences:
            random.shuffle(g2p)

        # Now we have a new preferences lists
        matching = stableMatching(n, group1Preferences, group2Preferences)
        #print (matching)

        #Check is stable
        if isStable(n, matching, group1Preferences, group2Preferences):
            countStableMatches = countStableMatches +1

    return (countStableMatches / 1000)*100  # Percentage
```
